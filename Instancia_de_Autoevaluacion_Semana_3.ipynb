1. **Cargar el conjunto de datos en una estructura adecuada para su manipulación.**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carga de datos
df_auto = pd.read_csv("/content/drive/MyDrive/Aprendizaje Automático/Archivos semana 3/Automobile.csv")
# Muestra de primeras filas
print("Primeras filas")
df_auto.head()
```

*Muestro Información General y Estadísticas descriptivas*

```python
# Información General
print("Información General")
df_auto.info()

# Estadísticas descriptivas
print("Estadísticas descriptivas")
df_auto.describe()
```

2.   **Crear un gráfico de dispersión para visualizar las relaciones entre los pares de características numéricas `mpg` y `weight`.**

```python
# Crear un gráfico de dispersión
sns.scatterplot(x='weight', y='mpg', data=df_auto, color ='seagreen')
plt.title('Relación entre Peso y Consumo de Combustible')
plt.xlabel('Peso del automóvil (weigth)')
plt.ylabel('Millas por galón (mpg)')

plt.show()
```

*Este gráfico de dispersión, nos muestra como el peso del automóvil influye en el consumo de combustible.*

*La correlación es negativa. A mayor peso (weight), el rendimiento (mpg) disminuye notablemente. Autos mas pesados, consumen mas combustible.*

3.   **Generar un histograma del consumo de combustible (`mpg`) para analizar su distribución.**

```python
# Histograma

plt.figure(figsize=(8, 5))
plt.hist(df_auto['mpg'], bins=20, color='seagreen', edgecolor='black')
plt.title('Distribución del Consumo de Combustible')
plt.xlabel('Millas por galón (mpg)')
plt.ylabel('Frecuencia')
plt.grid(axis='y')
plt.show()
```

*El histograma nos muestra la distribución del consumo de combustible (mpg).*

*Se puede observar una distribución asimétrica hacia la derecha (sesgo positivo), donde la mayoría de los valores se concentran entre 15 y 30 mpg.*

*Se observan también, algunos valores atípicos con consumos superiores a 40 mpg, lo que puede indicar la precencia de automóviles mas eficientes.*


4.   **Graficar un diagrama de caja (boxplot) de las características `cylinders` y `origin` para identificar posibles valores atípicos.**

```python
plt.figure(figsize=(12, 6))
df_auto.boxplot(column='cylinders', by='origin', grid=True)

plt.title('Distribución de los cilindros por origen')
plt.suptitle('')
plt.xlabel('Origen del automóvil')
plt.ylabel('Número de cilindros')
plt.show()
```

*El diagrama de caja muestra claramente las diferencias en la cantidad de cilindros según el origen de los automóviles. Europa y Japón muestran una fuerte concentración en motores de 4 cilindros, con poca variabilidad, lo que refleja un enfoque en la eficiencia. También, en EEUU y Japón, se observan algunos valores atípicos. En cambio, los automóviles de EEUU muestran una mayor dispersión, con valores que van entre 4 y 8 cilindros, y una mediana mas elevada, indicando una tendencia hacia motores mas potentes.*

5.   **Realizar un gráfico de barras que muestre la distribución de automóviles según su año de modelo (`model_year`) y origen (`origin`).**

```python
# Agrupamos los datos para contar cuántos autos hay por año y origen
counts = df_auto.groupby(['model_year', 'origin']).size().unstack()

# Graficar

counts.plot(kind='bar', figsize=(12, 6), color=['steelblue', 'pink', 'seagreen'])
plt.title('Distribución de Automóviles por Año de Modelo y Origen')
plt.xlabel('Año del modelo')
plt.ylabel('Cantidad de automóviles')
plt.legend(title='Origen', loc='upper right')
plt.grid(axis='y')
plt.show()
```

*El gráfico de barras muestra la preponderancia de los automóviles con origen estadounidense, especialmente en la década de 1970.*

*Europa muestra una presencia relativamente estable a lo largo del tiempo analizado, mientras que Japón muestra un creciemiento hacia finales de la década. Este cambio podría sugerir una transición en el mercado.*
