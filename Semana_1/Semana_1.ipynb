Ejercicio 7: Multiplica dos matrices 3x3 de enteros aleatorios entre 1 y 10.

```python
m1 = np.random.randint(1, 11, (3, 3))
m2 = np.random.randint(1, 11, (3, 3))

# Multiplicación matricial (producto punto)
resultado_dot = np.dot(m1, m2)
# O alternativamente: resultado_dot = m1 @ m2

print("Matriz 1:\n", m1)
print("Matriz 2:\n", m2)
print("Resultado:\n", resultado_dot)
```

Ejercicio 8: Crea un array de 15 valores aleatorios entre 1 y 100 e imprime su máximo, mínimo y promedio.

```python
datos = np.random.randint(1, 101, 15)

print(f"Array: {datos}")
print(f"Máximo: {datos.max()}")
print(f"Mínimo: {datos.min()}")
print(f"Promedio: {datos.mean()}")
```

Ejercicio 9: Genera una matriz 5x5 con valores del 1 al 25 y extrae su diagonal.

```python
# Creamos del 1 al 25 y redimensionamos a 5x5
matriz_5x5 = np.arange(1, 26).reshape(5, 5)
diagonal = np.diag(matriz_5x5)

print("Matriz:\n", matriz_5x5)
print("Diagonal:", diagonal)
```

Ejercicio 10 : Dada una matriz 5x5 de enteros aleatorios entre 1 y 20, calcula la suma de cada fila y columna

```python
matriz_rand_20 = np.random.randint(1, 21, (5, 5))

suma_columnas = matriz_rand_20.sum(axis=0)
suma_filas = matriz_rand_20.sum(axis=1)

print("Matriz:\n", matriz_rand_20)
print("Suma por columnas:", suma_columnas)
print("Suma por filas:", suma_filas)
```
